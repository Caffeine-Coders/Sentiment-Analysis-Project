```python
import numpy as np
import tensorflow as tf
from transformers import RobertaTokenizerFast
tokenizer = RobertaTokenizerFast.from_pretrained("roberta-base")
from transformers import TFRobertaForSequenceClassification
from tensorflow.keras.optimizers.legacy import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import load_model

from tensorflow.keras.callbacks import ModelCheckpoint
early_stop = EarlyStopping(monitor='val_sparse_categorical_accuracy', patience=3, verbose=3)

model = TFRobertaForSequenceClassification.from_pretrained("roberta-base", num_labels = 3)
model.compile(optimizer=Adam(3e-5), metrics = [tf.keras.metrics.SparseCategoricalAccuracy()])
```

```python
import re

def remove_mentions_and_links(text):
    # Remove @ mentions
    text = re.sub(r'@\w+', '', text)
    # Remove links
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'www\S+', '', text)
    return text
```
